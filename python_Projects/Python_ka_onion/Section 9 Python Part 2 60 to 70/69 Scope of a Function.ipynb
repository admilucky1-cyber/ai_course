```python
################
#################Function Inside Function###########

#As explained in the example above, the variable x is not available outside the function, but it is available for any function inside the function:
```

```python
'''Example
The local variable can be accessed from a function within the function:'''

def myfunc():
  x = 300             ## local variable of myfunc
  def myinnerfunc():
    print(x)            ## # accessing x
  myinnerfunc()

myfunc()
```

```python
'''üîç Explanation (Simple & Clear)

x = 300 is a local variable of the outer function myfunc()

myinnerfunc() is inside myfunc()

Python allows an inner function to access variables of its outer function

This is called Enclosing Scope (or closure)

üß† Scope Order in Python (LEGB Rule)

Python searches for variables in this order:

L ‚Üí Local (inside the function)

E ‚Üí Enclosing (outer function)

G ‚Üí Global

B ‚Üí Built-in

üëâ Here, x is found in the Enclosing scope, so print(x) works.

üìå Key Point (Exam Ready)

A local variable of an outer function can be accessed by an inner function.

‚ùó Important Note

The inner function can read the variable, but cannot modify it unless nonlocal is used.'''

def myfunc():
    x = 300
    def myinnerfunc():
        nonlocal x
        x = 500
    myinnerfunc()
    print(x)

myfunc()           #Output:500
```
