```python
phone = 'iphone10_onroad'


def my_home():
    phone = 'iphone10_onroad'
    return phone


print(my_home())
print(phone)
```

```python
#‚úÖ Correct Way 1: Using global keyword
phone = 'iphone10_onroad'

def parents_home():
    global phone
    phone = 'iphone10_onHome'

parents_home() 
print(phone)


#üìå Now the phone is really at home
```

```python
phone = 'iphone10_onroad'


def parents_home():
    phone = 'iphone10_onparents_home'
    return phone


print(parents_home())
print(phone)
```

```python
phone = 'iphone10_onroad'

def parents_home():
    return 'iphone10_onHome'

phone = parents_home()
print(phone)
```
