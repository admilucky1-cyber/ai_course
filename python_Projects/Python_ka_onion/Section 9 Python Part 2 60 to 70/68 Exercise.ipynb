```python

string = "a quick BroWnn Fox Jumps Over the Lazy Dog"
#word_count=len(string.split()) #to count total words
#print('Total words',word_count)#to print total words
############to count total letter::
#letters_count=len(string)
#print('Total Characters',letters_count)
counter = {'Upper_case_letter':0,'Lower_case_letter':0}

for letter in string:
  if letter.isupper():  
    counter['Upper_case_letter'] +=1
  elif letter.islower():
      counter['Lower_case_letter'] +=1
  else:
      pass
print(f'Th original String is :: \n{string}')
print(f'No. of Upper case letter: {counter["Upper_case_letter"]}')
print(f'No. of Lower case letter: {counter["Lower_case_letter"]}')
```
