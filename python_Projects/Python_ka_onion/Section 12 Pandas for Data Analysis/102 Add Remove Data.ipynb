```python
import pandas as pd
```

```python
phone_csv=pd.read_csv('cleaned_data.csv')
```

```python
phone_csv
```

```python
battries = pd.Series([3,4,5,6,7])
phone_csv['battery'] = battries
```

```python
phone_csv
```

```python
phone_csv['battery'] = phone_csv['battery'].fillna(5)
#This creates a new Series with missing values filled and assigns it back to the column.
#Clean, explicit, and avoids chained assignment issues.
phone_csv
```

```python
ringtones = [1,2,5,6,8,9,5,10,2,7]
phone_csv['Ringtones'] = ringtones
```

```python
phone_csv
```

```python
phone_csv['Product'] = phone_csv['battery']*phone_csv['Ringtones']
```

```python
phone_csv
```
