# This is a 'safety' step if your memory column ever contains text 
The Fix: Convert String to Float
# Remove the $ symbol and any commas, then convert to float
phone_csv["Price"] = phone_csv["Price"].str.replace(r'[\$,]', '', regex=True).astype(float)

# Now your command will work perfectly!
phone_csv[phone_csv['Price'] > 450]
## Whatâ€™s happening under the hood?
.str.replace(r'[\$,]', '', regex=True): This looks for both the dollar sign and any commas (in case a price is like $1,000) and deletes them.

.astype(float): This changes the data type from an "Object" (string) to a "Float" (decimal number).

1. Selecting a Single Cell with .iloc
The i in .iloc stands for Integer. It uses numbers for both the row and the column, starting from 0.

Formula: dataframe.iloc[row_index, column_index]

```python
# from gimi Select the 'Memory(kb)' (column 2) for the 4th row (index 3)
memory_value = phone_csv.iloc[3, 2]
print(f"The memory of the 4th phone is: {memory_value}")
```

2. Selecting a Single Cell with .loc
While .iloc uses numbers, .loc uses Labels. This is usually safer if you aren't sure which column number is which.

Formula: dataframe.loc[row_index, "Column_Name"]

```python
# from gimi__Select the 'Price' for the row where index is 2
price_value = phone_csv.loc[2, "Price"]
print(f"The price of the 3rd phone is: {price_value}")
```
