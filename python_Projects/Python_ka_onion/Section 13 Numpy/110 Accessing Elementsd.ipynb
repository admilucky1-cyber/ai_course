```python
a3d = np.array([[[1,2,3,3],[4,5,6,5]],[[1,2,3,5],[7,8,9,6]]])
```

```python
a3d
```

```python
a3d[:3]
```

```python
a3d[:2]
```

```python
a3d[:2,:1,:2]
# English: a3d[:2, :1, :2] means: first 2 blocks, first row only, first 2 columns.
# Roman Urdu: a3d[:2, :1, :2] ka matlab hai: pehle 2 blocks, sirf pehli row, aur pehle 2 columns.
#This is a great moment to reinforce axis-based slicing in 3D arrays.
```

# English: In 3D arrays, each block is a 2D matrix. In 4D arrays, each block is a 3D cube.
# Roman Urdu: 3D arrays mein har block ek 2D matrix hota hai. 4D arrays mein har block ek 3D cube hota hai.
# Rows aur columns block ke andar hote hain.
# English:
# Axis 0 → Block (a 3D cube)
# Axis 1 → Matrix inside the block
# Axis 2 → Row inside the matrix
# Axis 3 → Column inside the row

# Roman Urdu:
# Axis 0 → Block ya cube
# Axis 1 → Matrix block ke andar
# Axis 2 → Row matrix ke andar
# Axis 3 → Column row ke andar
