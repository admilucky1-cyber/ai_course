# A NumPy array is the fundamental data structure in the NumPy library, designed for efficient numerical computation. Since your notebook section is titled ‚Äú105 Numpy Arrays‚Äù, here‚Äôs a clear breakdown of what‚Äôs most relevant:

# üîé What a NumPy Array Is
It‚Äôs a grid of values (all of the same type) stored in contiguous memory.

Can be 1D (vector), 2D (matrix), or multi-dimensional.

Provides fast operations compared to Python lists because it‚Äôs implemented in C.

# ‚úÖ Key Features
Homogeneous data: All elements must be of the same type (e.g., all integers, all floats).

Shape: Defined by the number of dimensions and size of each dimension.

Example: arr.shape ‚Üí (3, 4) means 3 rows, 4 columns.

Efficient operations: Vectorized math (like arr * 2) applies element-wise without loops.

Indexing & slicing: Works like lists but more powerful, especially with multi-dimensional arrays.
