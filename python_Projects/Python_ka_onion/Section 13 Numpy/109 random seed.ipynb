#  üëç ‚Äî if you create a random array and want your teammate to see the same result, you need to set a seed before generating it. The seed ensures that the random number generator produces the same sequence every time, regardless of who runs the code or where it‚Äôs run.

# The seed() function in NumPy (and random number generation in general) is used to set the starting state of the random number generator. By fixing this starting point, you make the sequence of random numbers predictable and reproducible.

How it works
np.random.seed(42) initializes the random generator with the seed value 42.

Every time you run code after setting the same seed, you‚Äôll get the same random numbers.

Different seeds (e.g., 1, 7, 123) produce different sequences, but each seed always produces its own consistent sequence.

```python
#Seed
np.random.seed(seed=0)
random_matrix = np.random.rand(3,5)   #Auto range 0-1 
```
