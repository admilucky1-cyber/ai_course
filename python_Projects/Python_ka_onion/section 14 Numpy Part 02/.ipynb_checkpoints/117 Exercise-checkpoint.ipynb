# Step 2: Transpose (.T)
Transpose flips rows ‚Üî columns.
So your array becomes:

1	5
2	6
3	7

```python
a2d.T.shape 
#Transposed array (a2d.T)
# Transpose means flip rows and columns.
```

```python
a2d.T 
#üëâ Easy way to explain:
# Transpose is like rotating the table so that rows become columns and columns become rows.
```

# 1. What is reshape?
In NumPy, reshape changes the shape (rows √ó columns) of an array without changing the data.

Think of it like rearranging chairs in a classroom:

Same number of chairs (data).

Different arrangement (rows and columns).

```python
#Rehsape
a2d.reshape(2,3,1)
```

```python
a2d.reshape(2,3,1).shape
```

# 117. EXERCISE

```python
np.random.seed(seed=0)
item_amount = np.random.randint(20,size=(5,3))
item_amount
```

```python
item_price = np.array([12,8,11])
item_price.shape
```

```python
item_price_df = pd.DataFrame(item_price.reshape(1,3),
                             index=['item_price'],
                             columns=['item1','item2','item3'])
```

```python
item_price_df.shape,item_amount.T.shape
```

```python
total_price = item_price.dot(item_amount.T)
```

```python
total_price
```

```python
price = pd.DataFrame(total_price,index=['Mon','Tue','Wed','Thu','Fri'],columns=['item_price'])
price
```

```python
weekly_sales = pd.DataFrame(item_amount,index=['Mon','Tue','Wed','Thu','Fri'],columns=['item1','item2','item3'])
```

```python
weekly_sales
```

```python
weekly_sales['Total']=price
```

```python
weekly_sales
```
