# Step 2: Transpose (.T)
Transpose flips rows ‚Üî columns.
So your array becomes:

2	6
3	7

# 1. What is reshape?
In NumPy, reshape changes the shape (rows √ó columns) of an array without changing the data.

Think of it like rearranging chairs in a classroom:

Same number of chairs (data).

Different arrangement (rows and columns).

```python
#Rehsape
a2d.reshape(2,3,1)
```

```python
a2d.reshape(2,3,1).shape
```
