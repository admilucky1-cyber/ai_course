# 131.Plotting from Pandas Data Frame

```python
phone_data
```

```python
phone_data['Total Price'] = phone_data['Price'].cumsum()
```

```python
phone_data
```

```python
#above output is not as we require because price column data type is string we have to change it first int integer form
```

```python
phone_data['Total Price'] = phone_data['Price'].astype(int).cumsum()
```

```python
phone_data
```

```python
#Now look aboove output  is fulfilled
```

```python
phone_data['Date'] = pd.date_range('01/01/2026',periods=len(phone_data))
phone_data
```
