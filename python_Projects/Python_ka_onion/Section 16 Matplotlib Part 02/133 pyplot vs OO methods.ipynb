```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart = pd.read_csv('heart_disease_300.csv')
heart
```

```python
heart.head()
```

```python
over_40 = heart[heart['age']>40]
over_40
```

```python
over_40.plot(kind='scatter',
            x='age',
            y='chol',   #cholestrol
            c='target');
```

```python
# Object Oriented approach

fig,ax = plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',
            x='age',
            y='chol',   #cholestrol
            c='target',
            ax=ax);
```
