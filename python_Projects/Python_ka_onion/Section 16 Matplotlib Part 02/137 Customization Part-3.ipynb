# 135 Life Cycle of OO method Advanced

```python
#create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                 sharex=True)

# --- First Plot ---
scatter_plot = plot1.scatter(x=over_40['age'],
                             y=over_40['chol'],
                             c=over_40['target'])

plot1.set(title='Heart Disease Problem with Cholestrol',
          xlabel='Age',
          ylabel='Cholestrol')

plot1.legend(*scatter_plot.legend_elements(),
             title='Survived or Not',
             loc="upper left")   # position set here i did this loc seet hahaha

# --- Second Plot ---
scatter_plot = plot2.scatter(x=over_40['age'],
                             y=over_40['trestbps'],
                             c=over_40['target'])

plot2.set(title='Heart Disease Problem with resting Blood Pressure',
          xlabel='Age',
          ylabel='Resting Blood Pressure')

plot2.legend(*scatter_plot.legend_elements(),
             title='Survived or Not',
             loc="upper left");   # position set here

```

```python
import matplotlib.pyplot as plt

# Create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                 sharex=True)

# --- First Plot ---
scatter_plot = plot1.scatter(x=over_40['age'],
                             y=over_40['chol'],
                             c=over_40['target'])

plot1.set(title='Heart Disease Problem with Cholestrol',
          xlabel='Age',
          ylabel='Cholestrol')

plot1.legend(*scatter_plot.legend_elements(),
             title='Survived or Not',
             loc="upper left",
             bbox_to_anchor=(1,1))   # legend outside top-right

# Mean line
plot1.axhline(y=over_40['chol'].mean(), linestyle='--')

# --- Second Plot ---
scatter_plot = plot2.scatter(x=over_40['age'],
                             y=over_40['trestbps'],
                             c=over_40['target'])

plot2.set(title='Heart Disease Problem with resting Blood Pressure',
          xlabel='Age',
          ylabel='Resting Blood Pressure')

plot2.legend(*scatter_plot.legend_elements(),
             title='Survived or Not',
             loc="upper left",
             bbox_to_anchor=(1,1))   # legend outside top-right

# Mean line
plot2.axhline(y=over_40['trestbps'].mean(), linestyle="--")

plt.tight_layout()
plt.show()
```

```python
#create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                sharex=True
                                )
#Add data to plot1
scatter_plot = plot1.scatter(x=over_40['age'],
                           y=over_40['chol'],
                           c=over_40['target'])
# First get handles and labels from the scatter plot
handles, labels = scatter_plot.legend_elements()


#customize
plot1.set(title='Heart Disease Problem with Cholestrol',
         xlabel='Age',
         ylabel='Cholestrol',)
# Then pass them to legend(), and set the position there
plot1.legend(handles, labels, loc="upper left", title="Survived or Not") #this line was added by me because legend was getting random loction in graph so set to upper left corner

#plot1.legend(*scatter_plot.legend_elements(),title='Survived or Not')

####Now lets Plot Second Plot
scatter_plot = plot2.scatter(x=over_40['age'],
                           y=over_40['trestbps'],
                           c=over_40['target'])
#Customize
plot2.set(title='Heart Disease Problem with resting Blood Pressure',
         xlabel='Age',
         ylabel='Resting Blood Pressure')
plot2.legend(*scatter_plot.legend_elements(),title='Survived or Not');
```

```python
#create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                 sharex=True)

# Add data to plot1
scatter_plot = plot1.scatter(x=over_40['age'],
                             y=over_40['chol'],
                             c=over_40['target'],
                             cmap='Dark2')

# Customize
plot1.set(title='Heart Disease Problem with Cholestrol',
          xlabel='Age',
          ylabel='Cholestrol')

# Legend for plot1
handles, labels = scatter_plot.legend_elements()
plot1.legend(handles, labels, loc="upper left", title='Survived or Not')

#### Now lets Plot Second Plot
scatter_plot = plot2.scatter(x=over_40['age'],
                             y=over_40['trestbps'],
                             c=over_40['target'],
                             #cmap='Reds',
                             cmap='Set1'
                            )

# Customize
plot2.set(title='Heart Disease Problem with resting Blood Pressure',
          xlabel='Age',
          ylabel='Resting Blood Pressure')

# Legend for plot2
handles, labels = scatter_plot.legend_elements()
plot2.legend(handles, labels, loc="upper left", title='Survived or Not');
```
