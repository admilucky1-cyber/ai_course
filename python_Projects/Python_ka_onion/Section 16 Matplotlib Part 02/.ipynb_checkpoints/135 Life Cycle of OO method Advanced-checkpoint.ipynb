```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart = pd.read_csv('heart_disease_300.csv')
heart
```

```python
heart.head()
```

```python
over_40 = heart[heart['age']>40]
over_40
```

```python
over_40.plot(kind='scatter',
            x='age',
            y='chol',   #cholestrol
            c='target');
```

```python
# Object Oriented approach

fig,ax = plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',
            x='age',
            y='chol',   #cholestrol
            c='target',
            ax=ax);
```

```python
#customization of above
#create Figure and graph area
fig, ax = plt.subplots(figsize=(10,6))
#Home,aur usky andar room
#Plot the data
scatter_plot = ax.scatter(x=over_40['age'],
                         y=over_40['chol'],
                         c=over_40['target'])
#customization
ax.set(title='Suspectability of Heart Disease',
      xlabel='Age',
      ylabel='cholestrol') #Suspectability = Mutasir hony waly


#set the legend
ax.legend(*scatter_plot.legend_elements(), title='target')
scatter_plot;
```

# 135 Life Cycle of OO method Advanced

```python
#create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                sharex=True
                                )
#Add data to plot1
scatter_plot = plot1.scatter(x=over_40['age'],
                           y=over_40['chol'],
                           c=over_40['target'])
# First get handles and labels from the scatter plot
handles, labels = scatter_plot.legend_elements()


#customize
plot1.set(title='Heart Disease Problem with Cholestrol',
         xlabel='Age',
         ylabel='Cholestrol',)
# Then pass them to legend(), and set the position there
plot1.legend(handles, labels, loc="upper left", title="Survived or Not") #this line was added by me because legend was getting random loction in graph so set to upper left corner

#plot1.legend(*scatter_plot.legend_elements(),title='Survived or Not')

####Now lets Plot Second Plot
scatter_plot = plot2.scatter(x=over_40['age'],
                           y=over_40['trestbps'],
                           c=over_40['target'])
#Customize
plot2.set(title='Heart Disease Problem with resting Blood Pressure',
         xlabel='Age',
         ylabel='Resting Blood Pressure')
plot2.legend(*scatter_plot.legend_elements(),title='Survived or Not');
```
