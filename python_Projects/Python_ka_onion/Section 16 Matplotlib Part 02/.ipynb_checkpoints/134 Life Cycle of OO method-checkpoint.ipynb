```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart = pd.read_csv('heart_disease_300.csv')
heart
```

```python
heart.head()
```

```python
over_40 = heart[heart['age']>40]
over_40
```

```python
over_40.plot(kind='scatter',
            x='age',
            y='chol',   #cholestrol
            c='target');
```

```python
# Object Oriented approach

fig,ax = plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',
            x='age',
            y='chol',   #cholestrol
            c='target',
            ax=ax);
```

# 134 Entire Life Cycle of Object Oriented method

```python
#create Figure and graph area
fig, ax = plt.subplots(figsize=(10,6))
#Home,aur usky andar room
#Plot the data
scatter_plot = ax.scatter(x=heart['age'],
                         y=heart['chol'],
                         c=heart['target'])
scatter_plot;
```

```python
#customization of above
#create Figure and graph area
fig, ax = plt.subplots(figsize=(10,6))
#Home,aur usky andar room
#Plot the data
scatter_plot = ax.scatter(x=over_40['age'],
                         y=over_40['chol'],
                         c=over_40['target'])
#customization
ax.set(title='Suspectability of Heart Disease',
      xlabel='Age',
      ylabel='cholestrol') #Suspectability = Mutasir hony waly


#set the legend
ax.legend(*scatter_plot.legend_elements(), title='target')
scatter_plot;
```
