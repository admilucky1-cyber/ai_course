```python
# List Comprehension Case : 2
#whithout comments
'''li_two = []
for i in range(10):
    li_two.append(i**2)

print(li_two)


li_two = [i**2 for i in range(10)]
print(li_two)'''

# ============================================
# List Comprehension: Case 2
# ============================================

# Traditional way using for loop
# English: Create an empty list and append squares of 0‚Äì9
# Urdu: Khali list banai aur 0 se 9 tak numbers ke square add kiye
li_two = []
for i in range(10):
    li_two.append(i**2)

print(li_two)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# --------------------------------------------
# Compact way using list comprehension
# English: Same result in one line ‚Äî square each number
# Urdu: Wohi result ek line mein ‚Äî har number ka square
li_two = [i**2 for i in range(10)]
print(li_two)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]


```

```python
# ============================================
# List Comprehension: Case 3
# ============================================

# Traditional way using for loop with condition
# English: Create a list of squares of even numbers from 0‚Äì9
# Urdu: 0 se 9 tak ke even numbers ka square list mein add karein
li_three = []
for i in range(10):
    if i % 2 == 0:           # English: Check if number is even | Urdu: Dekhein kya number even hai
        li_three.append(i**2)  # English: Add square of even number | Urdu: Even number ka square add karein

print(li_three)  # Output: [0, 4, 16, 36, 64]

# --------------------------------------------
# Compact way using list comprehension
# English: Same logic in one line ‚Äî filter and transform
# Urdu: Wohi logic ek line mein ‚Äî filter aur transform dono
li_three = [i**2 for i in range(10) if i % 2 == 0]
print(li_three)  # Output: [0, 4, 16, 36, 64]


```

```python

# Hissa	       Code Part	        English Explanation	                      Roman Urdu Explanation
# Expression    i**2	             What to do with each item	          Har item ka square lena hai
# Loop	      for i in range(10) 	 Where items are coming from	      Kis list/range se values li ja rahi hain
# Condition	  if i % 2 == 0	        Which items to include (filtering)	  Sirf wo items jo even hain (2 se divide hotay hain)
```
