```python
# ============================================
# Python Special Function: reduce() 
# ============================================

# Step 1: Import reduce from functools
# English: reduce() is not built-in, we import it from functools
# Urdu: reduce() built-in nahi hai, isay functools se import karte hain
from functools import reduce

# Step 2: Sample data
# English: List of rainwater amounts
# Urdu: Barish ke paani ki miktar ki list
rain_water = [1, 2, 3, 4, 5]

# Step 3: Define a collector function
# English: This function adds two values together
# Urdu: Ye function do values ko jor deta hai
def collector(bucket, water):
    return bucket + water

# Step 4: Use reduce to accumulate total
# English: reduce() applies collector across the list, starting from 0
# Urdu: reduce() collector ko list ke har item par lagata hai, 0 se shuru kar ke
print(reduce(collector, rain_water, 0))  
# Output: 15

# --------------------------------------------
# Teaching Tip:
# English: Use reduce when you want one final result from a list
# Urdu: Jab aapko list se ek final result chahiye ho to reduce use karein
```
