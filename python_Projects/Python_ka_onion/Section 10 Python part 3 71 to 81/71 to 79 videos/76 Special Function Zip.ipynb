```python
'''1. How it works
Look at your code: you have a list of first names and a list of last names.

zip() takes the 1st person from the first list and matches them with the 1st person from the second list.

Then it does the same for the 2nd person, and so on.

The "Picture Method" (Zipper)
Imagine you are packing for a trip and matching socks:

Left Socks: ['Red', 'Blue']

Right Socks: ['Red', 'Blue']

Zip them together: [('Red', 'Red'), ('Blue', 'Blue')]

2. Breaking down your code
In your second example, you zipped three lists together:

First Names: 'Shahrukh', 'Anglina'

Last Names: 'Khan', 'julie'

Emails: 'Khan@blabla.com', 'julie@blabla.com'

The Result will look like this:



  ['Shahrukh', 'Khan', 'Khan@blabla.com'], # All the first items together
  ('Anglina', 'julie', 'julie@blabla.com') # All the second items together
]
Important Rule:
If one list is shorter than the others, zip() will stop at the shortest one.

If you have 10 first names but only 2 emails, the final list will only have 2 people in it. The other 8 names will be ignored.

3. Why use it?
It is very helpful when you have data in different lists that belongs to the same person. Instead of looking at three different lists, you use zip() to make a single list of complete profiles.'''
```

```python
# ============================================
# Python Special Function: zip()
# ============================================

# Two simple lists of numbers
# English: These are two lists we want to combine
# Urdu: Ye do lists hain jinhein hum jorna chahte hain
li_one = [1, 2]
li_two = [3, 4]

# Combine both lists using zip()
# English: zip() pairs elements by index from both lists
# Urdu: zip() dono lists ke elements ko unki position ke mutabiq jor deta hai
print(list(zip(li_one, li_two)))  
# Output: [(1, 3), (2, 4)]

# --------------------------------------------
# More practical example with names and emails
# --------------------------------------------

# First names list
# English: List of first names
# Urdu: Pehle naamon ki list
first_name = ['Shahrukh', 'Anglina']

# Last names list
# English: List of last names
# Urdu: Aakhri naamon ki list
last_name = ['Khan', 'julie']

# Emails list
# English: List of email addresses
# Urdu: Email addresses ki list
emails = ['Khan@blabla.com', 'julie@blabla.com']

# Combine all three lists
# English: zip() joins first name, last name, and email into one tuple per person
# Urdu: zip() har shakhs ke liye pehla naam, aakhri naam aur email ko ek tuple mein jor deta hai
print(list(zip(first_name, last_name, emails)))  
# Output: [('Shahrukh', 'Khan', 'Khan@blabla.com'), ('Anglina', 'julie', 'julie@blabla.com')]
```

```python
# ============================================
# Python Special Function: zip()
# ============================================

# First example: combining two lists
# English: Two lists of numbers
# Urdu: Do numbers ki lists
li_one = [1, 2]
li_two = [3, 4]

# English: zip() pairs elements by index from both lists
# Urdu: zip() dono lists ke elements ko unki position ke mutabiq jor deta hai
print(list(zip(li_one, li_two)))  
# Output: [(1, 3), (2, 4)]

# --------------------------------------------
# Second example: combining names and emails
# --------------------------------------------

# English: List of first names
# Urdu: Pehle naamon ki list
first_name = ['Shahrukh', 'Anglina']

# English: List of last names
# Urdu: Aakhri naamon ki list
last_name = ['Khan', 'julie']

# English: List of email addresses (only one email here)
# Urdu: Email addresses ki list (sirf ek email hai)
emails = ['julie@blabla.com']

# English: zip() stops at the shortest list — here, only one email
# Urdu: zip() sabse choti list tak hi kaam karta hai — yahan sirf ek email hai
print(list(zip(first_name, last_name, emails)))  
# Output: [('Shahrukh', 'Khan', 'julie@blabla.com')]

# --------------------------------------------
# Teaching Tip:
# English: Always check that all lists are equal in length before using zip()
# Urdu: zip() use karne se pehle lists ki length barabar honi chahiye
```
