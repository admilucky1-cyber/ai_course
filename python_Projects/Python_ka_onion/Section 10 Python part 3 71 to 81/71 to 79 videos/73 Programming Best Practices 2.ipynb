```python
# Quality of a Good Developer
# 1. Always return expected output
# 2. Avoid side effects (like printing inside return)

# Function definition
def adder(num):
    # Step: Add 1 to the input number
    # Roman Urdu: Input number mein 1 add kar rahe hain
    add_one = num + 1

    # Problematic: Returning the result of print(), which is None
    # Roman Urdu: Yeh print() return kar raha hai, jo None hota hai
    return print(add_one)

# Function call
# Roman Urdu: Function ko 4 ke input ke saath call kar rahe hain
adder(4)  # Output: 5 (printed), but return value is None
```
