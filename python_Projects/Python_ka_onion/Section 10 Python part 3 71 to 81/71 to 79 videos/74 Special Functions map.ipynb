```python
'''ðŸŒ± Final life lesson

Learn to work hard (loop)
Learn to work smart (map)
Grow wise by choosing correctly'''
# -------------------------------
# Example using map() function
# -------------------------------

# This is a list of numbers (think: student marks)
li = [1, 2, 3] #data

# This function works on ONE value at a time
def adder_map(x):
    # x is a single number coming from the list
    # add 1 to that number and return it
    return x + 1

# map() applies adder_map to each element of li
# list() collects the result into a list
result_map = list(map(adder_map, li))

print(result_map)

# Output:
# [2, 3, 4]


# -------------------------------
# Example using normal for-loop
# -------------------------------

# This function works on the WHOLE list
def adder_normal(para_li):
    # create an empty list to store new values
    new_li = []

    # loop through each value in the list
    for i in para_li:
        # add 1 to each value and store it
        new_li.append(i + 1)

    # return the final updated list
    return new_li

result_normal = adder_normal(li)

print(result_normal)

# Output:
# [2, 3, 4]


# -------------------------------
# Summary (Life Meaning)
# -------------------------------
# for-loop  -> manual work (step by step)
# map()     -> smart system (rule applied automatically)
# Both give the same result, choice depends on situation
```

```python
#map
#Same task using map() (shorter & cleaner)
#Since your comment mentions map, here is the same logic using map():

li = [1, 2, 3]

def add_one(x):
    return x + 1

result = list(map(add_one, li))
print(result)
```

```python
#Even shorter (using lambda)
li = [1, 2, 3]
print(list(map(lambda x: x + 1, li)))
```

```python
'''Comparison (important for exams)
Method	When to use
for loop	Best for beginners, very clear
map()	Functional style, shorter code
lambda + map	Very compact, less readable for beginner'''
```
