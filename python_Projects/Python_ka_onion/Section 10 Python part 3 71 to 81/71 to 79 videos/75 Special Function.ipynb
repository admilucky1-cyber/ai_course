```python
# ============================================
# Demonstration of Python's map() and filter()
# ============================================

# Our sample list of integers
li = [1, 2, 3, 4, 6, 8, 9]

# -------------------------------
# Using map()
# -------------------------------
# map() applies a function to every element in a list (or iterable).
# It returns an iterator, so we usually convert it to a list.
# Example: transform each number by adding 1.

def adder_map(para_li):
    # This function takes one number and adds 1
    return para_li + 1

# Apply adder_map to all elements of li
print(list(map(adder_map, li)))   # Output: [2, 3, 4, 5, 7, 9, 10]

# -------------------------------
# Using filter()
# -------------------------------
# filter() keeps only elements that satisfy a condition (True/False).
# It also returns an iterator, so we convert it to a list.
# Example: keep only odd numbers.

def odds(para_li):
    # This function returns True if the number is odd
    return para_li % 2 != 0

# Apply odds to all elements of li

print(list(filter(odds, li)))     # Output: [1, 3, 9]

# -------------------------------
# Extra prints
# -------------------------------
print(adder_map)   # Shows the function object (not very useful for beginners)
print(li)          # Original list remains unchanged
```

```python
# ============================================
# Python Special Functions: map() and filter()
# ============================================

# Original list of integers
li = [1, 2, 3, 4, 6, 8, 9]

# --------------------------------------------
# map() function
# --------------------------------------------
# Purpose: Apply a transformation to each element in a list.
# Here, we define a function that adds 1 to each number.

def adder_map(para_li):
    return para_li + 1

# Apply adder_map to each element using map()
mapped_result = list(map(adder_map, li))
print("Mapped Result (each +1):", mapped_result)

# Pros of map():
# ✅ Clean and readable for transformations
# ✅ Avoids explicit loops
# ✅ Efficient with large datasets
# Cons of map():
# ❌ Less intuitive for beginners
# ❌ Only works with one function at a time

# --------------------------------------------
# filter() function
# --------------------------------------------
# Purpose: Select elements that meet a condition.
# Here, we define a function that returns True for odd numbers.

def odds(para_li):
    return para_li % 2 != 0

# Apply odds to each element using filter()
filtered_result = list(filter(odds, li))
print("Filtered Result (odds only):", filtered_result)

# Pros of filter():
# ✅ Great for condition-based selection
# ✅ Cleaner than writing loops with if-statements
# ✅ Efficient and lazy evaluation
# Cons of filter():
# ❌ Only works with boolean-returning functions
# ❌ Chaining multiple filters can reduce readability

# --------------------------------------------
# Final check
# --------------------------------------------
print("Original List:", li)  # Confirm original data is unchanged
```

```python
# ============================================
# Python Special Function: filter()
# ============================================

# --------------------------------------------
# Filtering odd numbers from a list
# --------------------------------------------

# Sample list of integers
li = [1, 2, 3, 4, 6, 8, 9]

# Define a function that returns True for odd numbers
def odds(para_li):
    return para_li % 2 != 0  # True if number is odd

# Use filter() to keep only odd numbers
print("Odd numbers:", list(filter(odds, li)))  # Output: [1, 3, 9]

# Show original list to confirm it's unchanged
print("Original list:", li)

# --------------------------------------------
# Filtering names from a list
# --------------------------------------------

# Sample list of names
name = ["Sally", "John", "Cena", "Bradpitt", "Mohsin", "Asdsd", "IOon"]

# Define a function that excludes the name 'Cena'
def name_filter(para_name):
    if para_name != 'Cena':
        return para_name  # Any name except 'Cena' is returned

# Use filter() to exclude 'Cena'
print("Filtered names:", list(filter(name_filter, name)))  # Output: ['Sally', 'John', 'Bradpitt', 'Mohsin', 'Asdsd', 'IOon']

# --------------------------------------------
# Notes:
# --------------------------------------------
# ✅ filter() keeps elements where the function returns a truthy value (not just True).
# ✅ In the name_filter example, returning the name itself (a non-empty string) works like True.
# ❌ If the function returns None or an empty string, filter() will exclude that item.

# --------------------------------------------
# Teaching Tip:
# --------------------------------------------
# For clarity, it's better to return True/False explicitly:
# def name_filter(para_name):
#     return para_name != 'Cena'
```

```python
name = ["Sally", "John", "Cena", "Bradpitt", "Mohsin", "Asdsd", "IOon"]

# We add 'para_name' as a parameter so the function can 'see' the data
def name_filter(para_name):
    # We return True if we want to KEEP the name, False if we want to DROP it
    return para_name != 'Asdsd'

print("Filtered names:", list(filter(name_filter, name)))
```
