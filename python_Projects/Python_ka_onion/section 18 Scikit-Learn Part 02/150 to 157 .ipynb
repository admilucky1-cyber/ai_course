# 154 Getting Data Missing Values method 2
### Method 2 : Fill Nan values using sklearn

```python
phone_data_missing = pd.read_csv('phone_data_missing_values.csv')
phone_data_missing 
```

```python
phone_data_missing.isna().sum()
```

```python
phone_data_missing.dropna(subset=["Price"], inplace=True)
phone_data_missing.isna().sum()
```

```python
# Split X and y 
X = phone_data_missing.drop('Price',axis=1)
y = phone_data_missing['Price']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
import pandas as pd

# Column groups
cat_cols = ['Make', 'Colour']
num_cols = ['Sim Cards','Memory(kb)']

# Imputers
cat_imputer = SimpleImputer(strategy='constant', fill_value='missing')
num_imputer = SimpleImputer(strategy='mean')

# ColumnTransformer
imputer = ColumnTransformer(
    transformers=[
        ('cat', cat_imputer, cat_cols),
        ('num', num_imputer, num_cols)
    ]
)

# Apply imputation
phone_data_filled_X = imputer.fit_transform(X)

# Convert back to DataFrame
phone_data_filled_X = pd.DataFrame(
  phone_data_filled_X,
    columns=cat_cols + num_cols
)

phone_data_filled_X
```

# Lets try Ridge regression model

```python
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(1)

# Load California housing dataset
housing = fetch_california_housing()

# Build DataFrame
df_california = pd.DataFrame(housing.data, columns=housing.feature_names)
df_california["target"] = housing.target

# Split into train and test
X = df_california.drop("target", axis=1)
y = df_california["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Ridge regression
reg_model = Ridge()
reg_model.fit(X_train, y_train)

# Model score
print(reg_model.score(X_test, y_test))
```

```python
# to get better result or accuracy
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import pandas as pd
import numpy as np

# Reproducibility
np.random.seed(1)

# Load dataset
housing = fetch_california_housing()
df_california = pd.DataFrame(housing.data, columns=housing.feature_names)
df_california["target"] = housing.target

# Split data
X = df_california.drop("target", axis=1)
y = df_california["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Random Forest
rf_model = RandomForestRegressor(n_estimators=200, random_state=1)
rf_model.fit(X_train, y_train)
print("Random Forest R^2:", rf_model.score(X_test, y_test))

# Gradient Boosting
gb_model = GradientBoostingRegressor(n_estimators=200, random_state=1)
gb_model.fit(X_train, y_train)
print("Gradient Boosting R^2:", gb_model.score(X_test, y_test))
```
